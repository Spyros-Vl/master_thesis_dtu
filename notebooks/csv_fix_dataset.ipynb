```python
#MOVE THE TEST PICKLES TO ANOTHER FOLDER

random.seed(1234)
idx = random.sample(range(1000), 200)

for pickle in np.array(pickle_list)[idx]:
    shutil.move('../literature/Other/supervisely/wrist/pickle_data/'+pickle, '../literature/Other/supervisely/wrist/test_pickles/'+pickle)
```

```python
#MOVE SOME IMAGES FOR VALIDATION
pickle_list = os.listdir(root_dir + 'wrist/pickle_data')

random.seed(12)
idx = random.sample(range(800), 100)

for pickle in np.array(pickle_list)[idx]:
    shutil.move('../literature/Other/supervisely/wrist/pickle_data/'+pickle, '../literature/Other/supervisely/wrist/validation_pickles/'+pickle)
```

```python

random.seed(1234)
idx = random.sample(range(20327), 5000)

for img in np.array(img_dir)[idx]:
    shutil.move('literature/Other/supervisely/wrist/img/'+img, 'literature/Other/supervisely/wrist/test_images/'+img)

for annot in np.array(ann_dir)[idx]:
    shutil.move('literature/Other/supervisely/wrist/ann/'+annot, 'literature/Other/supervisely/wrist/test_annotations/'+annot)
```
