```python
#number of unique patients 
instances = df.filestem.values

tmp_list = []

for idx in range(len(instances)):
    text = instances[idx]
    tmp_list.append(text[:4])
```

```python
from collections import Counter

counts = Counter(tmp_list)
unique = list(counts.keys())
print(len(unique))
```

```python
print("We have " , len(unique) , " unique patients in this dataset" )
```

```python
#MOVE THE TEST PICKLES TO ANOTHER FOLDER

random.seed(1234)
idx = random.sample(range(1000), 200)

for pickle in np.array(pickle_list)[idx]:
    shutil.move('../literature/Other/supervisely/wrist/pickle_data/'+pickle, '../literature/Other/supervisely/wrist/test_pickles/'+pickle)
```
